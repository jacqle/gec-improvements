```python
import json
from pprint import pprint
```

```python
with open("data/FCE_errors.json","r") as f:
    obj = json.load(f)
obj = obj["code"]
```

```python
sort = sorted(obj.items(), key=lambda x: x[1]["frequency"], reverse=True)
sorted_dict = dict(sort)
```

```python
pprint(sorted_dict)
```

```python
print("\n\n\033[1m Most frequent errors \033[0m\n")
for key, value in sorted_dict.items():
    print(f"code: {key:<10} frequency: {value['frequency']:<15} Description: {value['explanation']:<40}")
```
